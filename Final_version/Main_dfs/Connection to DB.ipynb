```python
connection = pymysql.connect(host="localhost",user="root",passwd="",database="csv_db 6" )
cursor = connection.cursor()

cursor.execute('DROP TABLE IF EXISTS final_ranking')

sql="CREATE TABLE final_ranking(PID int, Ranking varchar(32), Zmiana_pozycji varchar(32), Miejsce int, Zawodnik varchar(32))"
cursor.execute(sql)

csv_data = csv.reader(open('final_ranking.csv',encoding="utf-8"))
next(csv_data)
for row in csv_data:
    cursor.execute('INSERT INTO final_ranking(PID, Ranking, Zmiana_pozycji, Miejsce, Zawodnik) VALUES(%s, %s, %s, %s, %s)',row)

connection.commit()
connection.close()
```

```python
connection = pymysql.connect(host="localhost",user="root",passwd="",database="csv_db 6" )
cursor = connection.cursor()

cursor.execute('DROP TABLE IF EXISTS table2')

sql="CREATE TABLE table_2(PID int, średnia_na_rozgrywce float, średnia_na_obronie float, Kokosy_na_Obronie float, częściówka float, końcówka float, szlemik float, Wpadka float, name varchar(32), klub varchar(32))" 

cursor.execute(sql)

csv_data = csv.reader(open('stats_to_website.csv',encoding="utf-8"))
next(csv_data)
for row in csv_data:
    cursor.execute('INSERT INTO table_2(PID, średnia_na_rozgrywce, średnia_na_obronie, Kokosy_na_Obronie, częściówka, końcówka, szlemik, Wpadka, name, klub) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)',row)

connection.commit()
connection.close()
```
